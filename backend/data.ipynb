```python
from time import sleep
import logging
from random import randint
from geopy.exc import GeocoderTimedOut, GeocoderServiceError

user_agent = 'user_me_{}'.format(randint(10000,99999))
geolocator = Nominatim(user_agent=user_agent)

sleep_sec=5



# exploring the location object
# attributes = [attr for attr in dir(location) if not attr.startswith('_')]

def extract_neighborhood(address):

    try:
        location = geolocator.geocode(address)
        
        if location is None:
            return 'Not Found'
        else:
            address_parts = location.address.split(',')

            if len(address_parts) >2:
                neighborhood = address_parts[2]
                return neighborhood.strip()
            
            else:
                return "Not found"
            
    except GeocoderTimedOut:
        logging.info('TIMED OUT: GeocoderTimedOut: Retrying...')
        sleep(randint(1*100,sleep_sec*100)/100)

    except GeocoderServiceError as e:
        logging.info('CONNECTION REFUSED: GeocoderServiceError encountered.')
        logging.error(e)
        return None
    except Exception as e:
        logging.info('ERROR: Terminating due to exception {}'.format(e))
        return None


```

```python
filtered_df['Neighborhood found'] = filtered_df['full_address'].apply(extract_neighborhood)

filtered_df
```

```python
def reverse_geocode(lat, lon):
    location = geolocator.reverse((lat, lon), exactly_one=True)
    address = location.raw['address']
    neighborhood = address.get('neighbourhood', '')

    return neighborhood
```

```python
filtered_df['Neighborhood'] = filtered_df.apply(
    lambda row: reverse_geocode(row['latitude'], row['longitude']) if pd.isnull(row['Neighborhood']) else row['Neighborhood'], axis=1
)
filtered_df
```
